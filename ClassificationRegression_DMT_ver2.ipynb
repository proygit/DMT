# Aplly K-fold cross validation - for Linear regression to predict male or female from input features Study Track,ML,IR,DB and statistics

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns

# select input features and target variable
X = df_csv_ver4[[  'taken DB before?','Gender','Used ChatGPT?']]
y = df_csv_ver4['taken ML before?']


# set up k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=0)

# initialize list to store MSE scores
mse_scores = []

# iterate over folds
for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    print(f"Fold {fold+1}")
    # split data into training, validation, and test sets
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_val, y_val = X.iloc[val_idx], y.iloc[val_idx]
    
    # fit a linear regression model on the training set
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # evaluate the model on the validation set
    y_pred = model.predict(X_val)
    mse = mean_squared_error(y_val, y_pred)
    mse_scores.append(mse)
    print(f"Mean Squared Error: {mse:.4f}")
    
# print the average MSE score across folds
print(f"Average MSE score: {sum(mse_scores)/len(mse_scores):.4f}")
# predict on the same data for visualization purposes
y_pred = model.predict(X)

```
